```python
import os
import numpy as np
import pandas as pd

input_dir = 'data'
```

```python
dfe = pd.read_csv(os.path.join(input_dir, 'enron_clean.csv'))
dfa = pd.read_csv(os.path.join(input_dir, 'spamassassin_clean.csv'))
```
